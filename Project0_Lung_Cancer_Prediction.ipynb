```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/cancer patient data sets.csv')
```

```python
df.head()
```

```python
#for i in range(0, 26):
#  print(f'{i} --> {df[df.columns[i]].dtype}')
```

```python
df1 = df.drop('Patient Id', axis=1)
```

```python
df1.head()
```

##Label Encoding

```python
from sklearn import preprocessing
labelencoder = preprocessing.LabelEncoder()
```

```python
df1['Level'] = labelencoder.fit_transform(df1['Level'])
```

```python
df1.head()
```

### Feature Selection (X & Y)

```python
#df1.columns
```

```python
features = ['Age', 'Gender', 'Air Pollution', 'Dust Allergy', 'OccuPational Hazards', 'Genetic Risk',
            'chronic Lung Disease', 'Smoking','Passive Smoker', 'Chest Pain', 'Coughing of Blood', 'Fatigue',
            'Weight Loss', 'Shortness of Breath', 'Wheezing', 'Clubbing of Finger Nails', 'Frequent Cold', 'Dry Cough',
            'Snoring']
x = df1[features]
y = df1['Level']
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y)
```

```python
from sklearn.svm import LinearSVC
linear_svc = LinearSVC()
linear_svc.fit(train_x, train_y)
```

```python
from sklearn.metrics import accuracy_score
```

```python
predictions = linear_svc.predict(test_x)
accuracy = accuracy_score(test_y, predictions)
```

```python
print(f' Prediction Accuracy: {accuracy * 100} %')
```
