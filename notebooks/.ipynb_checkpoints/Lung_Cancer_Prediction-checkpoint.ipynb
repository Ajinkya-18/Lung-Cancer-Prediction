# Lung Cancer Prediction

## Import Dependencies

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Data Loading and Preprocessing

```python
df = pd.read_csv('../data/cancer patient data sets.csv', index_col=0)
```

```python
df = df.drop(['Patient Id', 'Gender'], axis=1)
```

## Feature Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.feature_selection import mutual_info_regression, SelectKBest
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
ord_enc = OrdinalEncoder()
ord_enc.set_output(transform='pandas')
```

```python
scaler_cols = df.drop(['Level'], axis=1).columns.tolist()

transformers = [('standard scaler', StandardScaler(), scaler_cols)]

col_transformer = ColumnTransformer(transformers, 
                                    remainder='passthrough', sparse_threshold=0, n_jobs=6, 
                                    verbose=True, verbose_feature_names_out=False)
col_transformer.set_output(transform='pandas')
```

```python
X = df.drop(['Level'], axis=1)
Y = df['Level']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, 
                                                    random_state=42, shuffle=True)
```

```python
col_transformer.fit(x_train)
```

```python
# from joblib import dump
# with open('../models/col_transformer_fitted.joblib', 'wb') as f:
#     dump(col_transformer, f)
```

```python
ord_enc.fit(y_train.values.reshape(-1,1))
```

```python
# from joblib import dump 
# with open('../models/ordinal_encoder_fitted.joblib', 'wb') as f:
#     dump(ord_enc, f)
```

```python
x_train_new = col_transformer.transform(x_train)
x_test_new = col_transformer.transform(x_test)
y_train_new = ord_enc.transform(y_train.values.reshape(-1, 1))
y_test_new = ord_enc.transform(y_test.values.reshape(-1, 1))
```

## Model Selection and Training

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
svc = SVC()
rfc = RandomForestClassifier()
lr = LogisticRegression()
dtc = DecisionTreeClassifier()
```

```python
x_train_new.shape
```

```python
y_train_new.shape
```

```python
svc.fit(x_train_new, y_train_new.values.ravel())
rfc.fit(x_train_new, y_train_new.values.ravel())
lr.fit(x_train_new, y_train_new.values.ravel())
dtc.fit(x_train_new, y_train_new.values.ravel())
```

```python
from sklearn.metrics import accuracy_score, mean_squared_error
```

```python
svc_preds = svc.predict(x_test_new)
rfc_preds = rfc.predict(x_test_new)
lr_preds = lr.predict(x_test_new)
dtc_preds = dtc.predict(x_test_new)
```

```python
svc_acc = accuracy_score(y_test_new, svc_preds)
rfc_acc = accuracy_score(y_test_new, rfc_preds)
lr_acc = accuracy_score(y_test_new, lr_preds)
dtc_acc = accuracy_score(y_test_new, dtc_preds)
```

```python
print(f'SVC Accuracy: {svc_acc * 100} %')
print(f'RFR Accuracy: {rfc_acc * 100} %')
print(f'LR Prediction Accuracy: {lr_acc * 100} %')
print(f'DTC Prediction Accuracy: {dtc_acc * 100} %')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 15))

axs[0,1].scatter(y_test_new, svc_preds, color='green')
axs[0,1].set_title('SVC')

axs[1,0].scatter(y_test_new, rfc_preds, color='yellow')
axs[1,0].set_title('RFR')

axs[1,1].scatter(y_test_new, lr_preds, color='red')
axs[1,1].set_title('LR')


axs[0,0].scatter(y_test_new, dtc_preds, color='orange')
axs[0,0].set_title('DTC')

plt.show()
```

```python
from sklearn.metrics import confusion_matrix
lr_cm = confusion_matrix(y_test_new, lr_preds)
rfc_cm = confusion_matrix(y_test_new, rfc_preds)
svc_cm = confusion_matrix(y_test_new, svc_preds)
dtc_cm = confusion_matrix(y_test_new, dtc_preds)
```

```python
plt.title('Logistic Regression - Confusion Matrix')
sns.heatmap(lr_cm, annot=True, cmap='Blues', fmt='g')
plt.savefig('../reports/lr_cm.png')
plt.show()
```

```python
plt.title('Support Vector Machine - Confusion Matrix')
sns.heatmap(svc_cm, annot=True, cmap='Blues', fmt='g')
# plt.savefig('../reports/svc_cm.png')
plt.show()
```

```python
plt.title('Random Forest Regressor - Confusion Matrix')
sns.heatmap(rfc_cm, annot=True, cmap='Blues', fmt='g')
plt.savefig('../reports/rfr_cm.png')
plt.show()
```

```python
plt.title('Decision Tree Classifier - Confusion Matrix')
sns.heatmap(dtc_cm, annot=True, cmap='Blues', fmt='g')
# plt.savefig('../reports/dtc_cm.png')
plt.show()
```

```python
def get_value_range(df) -> tuple:
    features_range = {}
    for col in df.columns:
        features_range[col] = (min(df[col]), max(df[col]))

    return features_range
```

```python
def get_predictions(test_data, model, scaler, encoder):
    return encoder.inverse_transform(model.predict(scaler.transform(test_data)))
```

```python
print(f'Lung Cancer Susceptibility: {get_predictions(test_data, knc_loaded, std_scaler, label_encoder)[0]}')
```

```python
from shap import KernelExplainer, LinearExplainer, TreeExplainer, sample
```

```python
dt_explainer = TreeExplainer(dtc)
dt_shap_vals = dt_explainer.shap_values(x_train_new)
rf_explainer = TreeExplainer(rfc)
rf_shap_vals = rf_explainer.shap_values(x_train_new)
```

```python
svm_explainer = KernelExplainer(svc.predict, sample(x_train_new, 100))
svm_shap_vals = svm_explainer.shap_values(x_test_new)
```

```python
lr_explainer = LinearExplainer(lr, x_train_new)
lr_shap_vals = lr_explainer.shap_values(x_test_new)
```
